```python
def json_to_dataframe(data_entrada):
    elements = data_entrada['elements']
    places = {'lat': [], 'lon': [], 'name': [], 'address': []}

    for i in elements:
        lalitude = i['lat']
        longitude = i['lon']
        name = i['tags'].get('name')
        street = i['tags'].get('addr:street', 'Sin calle')
        number = i['tags'].get('addr:housenumber', 0)
        
        places['lat'].append(lalitude)
        places['lon'].append(longitude)
        places['name'].append(name)
        places['address'].append(str(street) + ' ' + str(number))

    df = pd.DataFrame(places)
    
    return df
```

#### Acceder a datos de la api desde la URL con la libreria requests

```python
# https://www.themoviedb.org/documentation/api
api_key = "api_key"

years = np.arange(2010, 2011)
id_movies = []
release_date = []
title = []
for i in years:
    print("year:", i )
    for j in range(1,5):
        response = requests.get(f'https://api.themoviedb.org/3/discover/ \
                                movie?api_key={api_key}&page={j}&primary_release_year={i}&sort_by=popularity.desc')
        highest_revenue = response.json()

        first_results = highest_revenue["results"]
       
        for k in first_results:
                     
            try:                
                release_date.append(k["release_date"])
                title.append(k["title"])
                id_movies.append(k["id"])  
            except:
                release_date.append(np.nan)
                title.append(np.nan)
                id_movies.append(np.nan)
                continue

df_movies_id = pd.DataFrame(id_movies, columns=["id"])
df_movies_id["date"] = release_date
df_movies_id["title"] = title
```

```python
df_movies_id
```

```python
df_movies_id["id"].values
```

```python
dict_movies = {"title":[], "revenue":[], "budget":[]}
for q in df_movies_id["id"].values[:11]:
    response = requests.get(f'https://api.themoviedb.org/3/movie/{q}?api_key=' +  api_key + "&append_to_response=keywords,credits")
    highest_revenue = response.json()
    try:
        budget = highest_revenue["budget"]
        revenue = highest_revenue["revenue"]
        title = highest_revenue['title']
        dict_movies["title"].append(title)
        dict_movies["revenue"].append(revenue)
        dict_movies["budget"].append(budget)
    except:
        continue
df_movies_2010 = pd.DataFrame(dict_movies)

```

```python
search_response = youtube.search().list(
  q="Chat gpt",
  part="snippet",
  order="viewCount",
  type="video",
  maxResults=50
).execute()

id_video = []
channelTitle = []
publish_time = []
description = []
title_video = []

for search_result in search_response.get("items", []):
    id_video.append(search_result["id"]["videoId"])
    channelTitle.append(search_result["snippet"]["channelTitle"])
    publish_time.append(search_result["snippet"]["publishTime"])
    description.append(search_result["snippet"]["description"])
    title_video.append(search_result["snippet"]["title"])


duration = []
definition = []
closed_caption = []
license_content = []
madeforkids = []
view_count = []
like_count = []
favorite_count = []
comment_count = []
license = []

for id_video_search in id_video:
    video_info_response = youtube.videos().list(
      part = ["contentDetails", "statistics","status"],
      id = id_video_search
    ).execute()
    duration .append(video_info_response["items"][0]['contentDetails']["duration"])
    definition.append(video_info_response["items"][0]['contentDetails']["definition"])
    closed_caption.append(video_info_response["items"][0]['contentDetails']["caption"])
    license_content.append(video_info_response["items"][0]['contentDetails']["licensedContent"])
    madeforkids.append(video_info_response["items"][0]['status']["madeForKids"])
    license.append(video_info_response["items"][0]['status']["license"])
    view_count.append(video_info_response["items"][0]['statistics']["viewCount"])
    like_count.append(video_info_response["items"][0]['statistics']["likeCount"])
    favorite_count.append(video_info_response["items"][0]['statistics']["favoriteCount"])
    comment_count.append(video_info_response["items"][0]['statistics']["commentCount"])
    

df_youtube = pd.DataFrame({"id_video": id_video,
                           "Channel": channelTitle,
                           "title" : title_video,
                           "description": description,
                           "publish_time": publish_time,
                           "duration":duration,
                           "definiton": definition,
                           "closed_caption":closed_caption,
                           "license_content": license_content,
                           "madeforkids":madeforkids,
                           "license": license,
                           "view_count": view_count,
                           "like_count":like_count,
                           "favorite_count" : favorite_count,
                           "comment_count": comment_count
                           })
df_youtube.head()
```

### Generador de Imágenes

```python
#Crear una imagen a partir de un texto (prompt)

response = openai.Image.create(
  prompt="a dog, a cat and a little chicken",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']
image = requests.get(image_url)

with open("data/dog_cat_chicken.jpg", "wb") as f:
    f.write(image.content)
```

```python
#Crea la variacion de una imagen (cats.png)

response = openai.Image.create_variation(
  image=open("data/cats.png", "rb"),
  n=1,
  size="1024x1024"
)

image_url = response['data'][0]['url']

image = requests.get(image_url)

with open("data/cats_variation.png", "wb") as f:
    f.write(image.content)
```

```python
# Completa una image a partir de la original (mona_lisa) y una mascara (mask)

response = openai.Image.create_edit(
  image=open("data/mona_lisa.png", "rb"),
  mask=open("data/mask.png", "rb"),
  prompt="a women with a baseball cap and sun glasses",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']

image = requests.get(image_url)

with open("data/new_picture.png", "wb") as f:
    f.write(image.content)
```
